## Crisis Communication: The Key To Good Brand Management

### The Impetus for Good Crisis Communication

   Crisis communications is crucial in maintaining brand integrity and consumer trust. The way companies navigate challenging times and react to crises is ever changing and has greatly evolved. How companies respond to issues like product recalls, safety concerns, and public backlash has changed significantly over the years. Ranging from public statements to proactive action, corporate responses reflect both the priorities companies have as well as the shift in societal expectations. In our research, we aim to understand how companies manage public sentiment during crises and how these strategies have changed over time. To do so, we examined decades of various corporate communications to uncover patterns, successes, and lessons learned, offering a perspective into the changing language and strategies of crisis management. 

   The 1982 Johnson & Johnson Tylenol recall remains one of the most defining moments in crisis communication history. In 1982, Johnson & Johnson faced a crisis that threatened its flagship product, Tylenol. A criminal tampered with the Extra Strength Tylenol capsules. Someone had opened the capsules and replaced the pain relieving medicine with deadly doses of potassium cyanide. One of the most common over the counter pain relievers had been turned into a murder weapon. After seven people in the Chicago area died from consuming cyanide laced Tylenol capsules, public panic erupted nationwide. Consumers feared for their safety and media coverage amplified concerns about the security of medications. The news of the deaths sparked fear in households nationwide, as Tylenol was one of the most trusted and widely used pain relievers. Pharmacies, grocery stores, and hospitals rapidly pulled Tylenol products from shelves, creating a domino effect of alarm in both the healthcare industry and retail sectors. The media broadcasted continuous coverage of the fatalities, leaving many Americans questioning the safety of Tylenol. Headlines showed warnings, talk shows debated corporate responsibility, and news broadcasts featured distraught family members, intensifying the atmosphere of dread.

   J&J’s (Johnson & Johnson’s) response to this crisis was groundbreaking. In the chaos, Johnson & Johnson had the spotlight, and needed a way to regain consumer trust and ensure safety. The company recalled 31 million bottles nationwide, halting production and sales, despite the immense financial loss. J&J launched a national communication campaign to inform the public, established a hotline for concerned consumers, and cooperated with law enforcement to investigate the tampering. J&J also introduced packaging that hindered the ability to tamper with, setting a new standard for the industry. This move demonstrated a prioritization of public safety over profits, a stance that resonated deeply with the public.

### The Point: Was There Marked Change?

   J&J was marked as very influential in the way companies handle crisis. Hence, the following hypothesis was developed to determine whether or not companies adhered to this standard: Following the famous J&J recall (which established a gold standard protocol by which companies could/should follow), the effectiveness of crisis communications within the pharmaceutical/drug industry should improve over time, in line with this standard. To do so, we examined corporate communication strategies across decades, analyzing their language, tone, and priorities to understand how approaches to crisis management have evolved.

### The Insights:
 
   Analysis of corporate crisis communication in the 1980s began with a detailed examination of a corpus of statements and media reports from the decade. Using computational text analysis, we tokenized and normalized the text to build frequency lists and identify recurring patterns in corporate messaging. Terms like “recall,” “capsules,” and “tests” appeared prominently, signaling a reactive focus on addressing immediate crises. Bigram and trigram analyses revealed that phrases such as “the company said” and “FDA said” were common, indicating a reliance on statements that deflected responsibility onto regulators or external entities rather than accepting accountability. Sentiment analysis further highlighted the defensive tone of much of the language, often emphasizing corporate image rather than public safety.
    
   We also uncovered notable discourse patterns through context analysis, which provided a closer look at how terms like “safety” and “responsibility” were used. Many statements lacked direct acknowledgments of corporate fault, reflecting ambiguity in accountability. Reports of test evasion and unethical practices emerged in the data, illustrating a broader trend of avoiding full transparency. These contextual findings show the corporate mindset of the 1980s, which was heavily focused on mitigating reputational damage rather than addressing root causes of crises or prioritizing consumer trust.

![](images/tyl_rec_3.png)

   In the 1990s, corporate crisis communication began to shift, with a growing focus on public health impacts and regulatory oversight. Frequency analysis of the top 100 words revealed a continued emphasis on recalls and safety concerns, with terms like “recall,” “patients,” and “doctors” frequently appearing. The prominence of words such as “Fen-Phen,” “Redux,” and “Fenfluramine” highlighted the weight loss drug controversies of the decade, which underscored the significant health risks posed by certain pharmaceutical products. Phrases like “the drug,” “the company,” and “the FDA” in bigram and trigram analysis underscored the central roles of corporations and regulatory bodies in addressing these crises. However, the data also revealed a defensive tone, with terms such as “he said” and “the company said” reflecting a reliance on statements aimed at mitigating reputational damage rather than prioritizing public accountability.
    
   Context analysis provided deeper insight into the framing of consumer concerns and trust. Phrases like “consumer safety” and “consumer protection” signaled an increased awareness of the need to address public health, yet these terms were often juxtaposed with defensive language emphasizing corporate image. The term “showed” frequently appeared in test results or studies, but its use was sometimes ambiguous, as phrases like “company tests showed they were of a relatively harmless” suggested a reluctance to fully acknowledge the severity of safety concerns. Additionally, public trust in corporations remained tenuous, with KWIC analysis of “trusted” highlighting skepticism about corporate responsibility for consumer health. The controversies surrounding Fen-Phen and Redux, particularly their association with heart valve damage, magnified this distrust and emphasized the need for clearer communication and accountability.
    
   Despite incremental progress in acknowledging health risks and regulatory responsibilities, corporate communication in the 1990s still fell short in many areas. Companies often framed their messaging around financial and reputational implications, as seen in the prominence of terms like “sales of,” “million,” and “market.” While the language began to reflect public health priorities, its defensive tone and ambiguous accountability revealed that companies were still learning how to balance transparency with self preservation. The persistent skepticism captured in the data highlighted the necessity for clearer, more proactive communication strategies to regain consumer trust and address the systemic challenges of the era.

![](images/90s_image.png)

   In the 2000s, corporate crisis communication demonstrated a notable shift toward prioritizing public health and patient safety. Our analysis revealed frequent mentions of key terms like “public health,” “patients,” and “safety,” signaling an increased focus on addressing consumer concerns. This period saw major recalls, including drugs like "Vioxx" and "Baycol," and heightened scrutiny of companies such as "Merck" and "Bayer." Bigrams like “the drug” and “risk of” underscored the growing emphasis on drug safety and associated risks, while trigrams such as “increased risk of” highlighted a more precise acknowledgment of health consequences. Alongside this, the frequency of terms like “compliance” and “quality assurance” reflected a greater commitment to regulatory standards and improved product safety.
    
   Looking again at context, insights within this realm revealed the evolving strategies in corporate messaging. Phrases such as “steps to ensure safety” and “updated safety information” illustrated companies’ efforts to proactively address safety concerns. Mentions of “public health” and “safety of medicines” in international contexts pointed to a more global approach to crisis management. Additionally, statements like “our first responsibility is to safety” marked an important shift in tone, emphasizing patient welfare as a priority. However, the data also showed that challenges persisted, with phrases like “risks outweigh safety” demonstrating the difficulty of balancing drug benefits and potential dangers. Companies also increasingly referenced “landmark studies” and “clinical trial data,” reflecting the need for rigorous research to build credibility and assure public confidence.
    
   While corporate communication in the 2000s marked a significant effort to rebuild trust, the data highlighted ongoing vulnerabilities. References to “consumer safety” and “trust building” showed companies’ attempts to strengthen consumer confidence, but phrases like “a breakdown of trust” revealed how past failures continued to undermine these efforts. Enhanced safety warnings and increased transparency were meaningful steps forward, but the persistent focus on financial and reputational implications suggested that companies were still navigating the balance between public accountability and economic priorities. Overall, the 2000s reflected a period of progress, as organizations increasingly prioritized compliance, safety, and public health, laying a foundation for more proactive strategies in the years to come.

   In the 2010s, corporate crisis communication underwent a transformation as public health and safety became central themes, driven by an era of heightened consumer awareness and regulatory scrutiny. In the 2010s, corporate crisis communication shifted to prioritize consumer safety and public health, driven by increasing regulatory scrutiny and heightened consumer awareness. Our analysis revealed frequent mentions of drugs like "Valsartan," "Ranitidine," and "Losartan," tied to significant recalls over contamination concerns, particularly involving NDMA impurities. Keywords such as "patients," "health," and "risk" highlighted the emphasis on consumer safety, while terms like "manufacturing," "impurity," and "quality" underscored the focus on production standards and failures. Bigram and trigram analyses revealed phrases like “blood pressure” and “trace amounts of,” reflecting both the health risks and the technical precision of recalls during this decade. Statements framed recalls as proactive measures, with phrases like “voluntary recall of” illustrating a shift toward accountability and risk mitigation.
    
   Context further illuminated how corporations treated the consumer as a central stakeholder. Mentions of "consumer safety" and "consumer complaints" demonstrated the increasing role of feedback in shaping corporate responses. Phrases like “commitment to health and safety” signaled efforts to reassure the public, while references to “consumer watchdog” and “consumer protection law” highlighted the growing regulatory focus on safeguarding consumers. However, the data also revealed challenges, with phrases such as “a breakdown of trust” and “responsible for safety” underscoring the struggle to maintain credibility. While corporate messaging often emphasized proactive safety measures, lapses in quality control and transparency occasionally undermined these efforts. The 2010s reflected a pivotal period in which consumers transitioned from passive recipients of corporate actions to active participants influencing crisis communication, signaling both progress and persistent challenges in maintaining public trust.

   In the 2020s, corporate crisis communication reflected a significant evolution toward precision, accountability, and proactive measures, driven by advancements in testing and regulatory scrutiny. Our analysis highlighted a focus on specific drugs like "Chantix," "Metformin," and "Hydrochloride," tied to recalls addressing contamination issues such as NDMA impurities. Keywords like "safety," "health," and "risk" underscored the continuing emphasis on consumer well being, while terms such as "adverse events" and "safety profiles" signaled a more technical approach. Bigram and trigram analyses revealed an emphasis on phrases like “due to” and “the recall,” indicating increased transparency in addressing safety concerns. Mentions of “acceptable daily intake” and “distributed across the US” reflected global scale and regulatory specificity, showing a broader geographic and technical scope compared to earlier decades.
    
   Context analyses of terms like “consumer” and “safety” highlighted a more personalized and trust centered approach to crisis communication. Phrases such as “consumer safety concerns” and “enhance public health and consumer welfare” showed a pivot toward addressing individual consumer impacts, moving beyond the more corporate driven narratives of the 1980s and 1990s. Meanwhile, “trust” was often framed in terms of empowering consumers through clearer information, with mentions like “trusted internet drug shop” emphasizing the growing importance of online reputation and verified services. Companies leaned into proactive messaging, with phrases such as “commitment to safety” and “out of caution after test results” signaling an anticipatory stance rather than reactive crisis management. The decade marked a continued evolution toward transparency, yet challenges remained in maintaining authenticity, as the growing complexity of regulatory and consumer expectations placed increasing demands on corporate communication strategies.


### Some More Insight...How do certain words appear specifically across decades?

1. `Safety`
Initially tied to immediate health concerns, safety now includes compliance with regulations and proactive measures. This evolution shows a shift from reactive crisis management to preventative responsibility, highlighting the growing importance of safeguarding public trust.

2. `Consumer`
The tone of corporate references to “consumers” has shifted significantly. In earlier decades, companies spoke to consumers as passive recipients of decisions. Today, there’s a more authentic approach, reflecting trust building efforts and a recognition of consumers as active stakeholders in corporate actions.

3. `Responsibility`
“Responsibility" has grown in importance, evolving from being reactive to proactive. However, its sincerity often appeared tied to public backlash, emphasizing the need for genuine accountability supported by meaningful actions.

4. `Trust`
The concept of “trust” has remained central but is fragile, particularly because social media amplifies criticism and backlash. While companies often attempt to rebuild trust with polished statements, only consistent transparency and action can ensure credibility over time.

### And Finally, Some Key Takeaways for Effective Crisis Communication:
* Transparency is Crucial: Honest communication fosters quicker recovery. J&J’s Tylenol recall remains the gold standard.
* Consistency Builds Trust: Mixed messages hurt credibility, making unified statements essential.
* Language Reflects Expectations: The shift to proactive language aligns with rising demands for corporate accountability.
* Trust Requires Time: Rebuilding trust demands sustained actions, not just crisis driven responses.
